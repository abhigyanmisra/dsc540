```python
# Step: Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step: Read XLS file from local directory
df_ff = pd.read_excel("13. PC Wise Voters Turn Out.xls", skiprows=3)
df_ff.head()
```

```python
# Task 1 - Replacing Headers
# MALE - Renamed to Male Vote Count
# FEMALE - Renamed to Female Vote Count
# THIRDGENDER - Renamed to ThirdGender Vote Count
# TOTAL is Sum of above three or sum of all EVM votes so renaming it Total EVM Votes Count
# Unnamed: 9 is Postal Votes Count - So will name column as Postal Votes
# TOTAL VOTERS - Renamed as Total Votes
# MALE.1 is renamed as MALE VOTES PERCENT
# FEMALE.1 is renamed as FEMALE VOTES PERCENT
# THIRDGENDER.1 is renamed as THIRDGENDER VOTES PERCENT

# Additionally - Also need to skip footer 5 rows, was having issues when trying to convert data type
header_ff = ["State Name", "PC No", "PC Name", "Electors","Male Vote Count","Female Vote Count", 
             "Third Gender Vote Count","Total EVM Votes Count","NRI Votes Count","Postal Votes Count","Total Votes Count", "Voter Turnout %",
            "Male Votes %", "Female Votes %","Third Gender Votes %"]

df_ff = pd.read_excel("13. PC Wise Voters Turn Out.xls", skiprows=3,skipfooter=5,names=header_ff)
df_ff
```

```python
# Task 2 - Checking for Outliers & Bad Data
# State Name, PC No, PC Name Should not be empty
print("Is Any State Name Null ? - %r " % df_ff['State Name'].isnull().values.any())
print("Is Any PC No Null ? - %r " % df_ff['PC No'].isnull().values.any())
print("Is Any PC Name Null ? - %r " % df_ff['PC Name'].isnull().values.any())

# Task 3 - Rounding % values to 2 decimal places for better readability
df_ff['Voter Turnout %'] = df_ff['Voter Turnout %'].round(decimals=2)
df_ff['Male Votes %'] = df_ff['Male Votes %'].round(decimals=2)
df_ff['Female Votes %'] = df_ff['Female Votes %'].round(decimals=2)
df_ff['Third Gender Votes %'] = df_ff['Third Gender Votes %'].round(decimals=2)
df_ff
```

```python
#Task 4 - Find duplicates
# To check for duplicates, PC Name Should be Unique (Generally it should be combination of PC Name and No)
print("Is Any PC Name duplicate ? - {}".format(any(df_ff['PC Name'].duplicated())))
duplicateRowsDF = df_ff[df_ff.duplicated(['PC Name'])]
display(duplicateRowsDF)
# Checking the Combination of PC Name and PC No
duplicateRowsDF = df_ff[df_ff.duplicated(['PC Name','PC No'])]
display(duplicateRowsDF)
# Since the Combination check returns no values we are good
```

```python
# Task 5 - Checking Validity and adding a column for Validity
# Logic Used for validity - 
# 1. Male Vote Count + Female Vote Count + Third Gender Vote Count= Total EVM Votes Count
# 2. Total EVM Votes Count + Postal Votes Count = Total Votes Count

# Case 1.
df_ff["Validity"] = (df_ff['Male Vote Count'] +
                     df_ff['Female Vote Count'] +
                     df_ff['Third Gender Vote Count'] == df_ff['Total EVM Votes Count'])

# FOR DEBUGGING - Uncomment if needed - print("Any Invalid Values due to Case 1 - ",(df_ff['Validity'] == False ).any())
# Case 2.
df_ff["Validity"] = (df_ff['Total EVM Votes Count'] +
                     df_ff['Postal Votes Count'] == df_ff['Total Votes Count'])

# FOR DEBUGGING - Uncomment if needed - print("Any Invalid Values due to Case 2 - ",(df_ff['Validity'] == False ).any())

display(df_ff)
```

```python
# Task 6 - Checking if all Data is valid
print("Is Any Data Invalid ? - {}".format(any(df_ff['Validity'] == False)))
```

```python
# importing required libraries
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Utility Functions to read the URL and save it as html in local to read later as needed
# Wrting the result to a file out.html

# URL of the website data. Data is available in tables.
wiki_url = "https://en.wikipedia.org/wiki/Results_of_the_2019_Indian_general_election"

def writeToHTML(result):
    print("Writing to File..Start")
    with open("out.html", 'wb') as f:
        f.write(result)
    print("Writing to File..Done")

import urllib.request

# Reading the HTML Page
with urllib.request.urlopen(wiki_url) as url:
    s = url.read()
    # Ithis would output the html source code
    writeToHTML(s)
```

```python
# Reading the downloaded file - out.html
with open("out.html", "r", encoding="utf-8") as fd1:
    soup = BeautifulSoup(fd1, "html.parser")
    print(type(soup))
```

```python
# Task: Find the tables you will need to deal with(how many tables are there?)
all_tables = soup.find_all("table")
print("Total number of tables are {} ".format(len(all_tables)))
```

```python
# Task: Find the right table using bs4 
# After inspecting the data, we find that the one with class=wikitable sortable collapsible" is the 
# table that holds
data_table = soup.findAll("table", {"class": "wikitable sortable collapsible"})
print(type(data_table))
print(len(data_table))
```

```python
# As we see there are two tables after that filtering, lets look at them and see which one is ours
# after inspection we see that the one of our interest is in second place at line number 4274
# while the other one is at line 450
# So Reading the one of our interest
header_web = [th.getText().strip() for th in data_table[1].findAll('tbody')[0].findAll('th')]
header_web
```

```python
# We may not be able to use headers directly as we have multiple levels for rows and also one present in column
# the values of interest lie between
# 'Margin', - this would be the last column, which is difference of votes between Winner and Runner Up
# 'No.', - this is Parliamentary Constituency No.
# 'Name', - This is Name of Parliamentary Constituency
# 'Type', - This is the Parliamentary Constituency Type
# 'Candidate', - This is Winning Candidate Name
# 'Party', - This is Winning Candidate Party
# 'Votes', - This is Winning Candidate Votes Count
# 'Candidate', - This is RunnerUp Candidate Name
# 'Party', - This is RunnerUp Candidate Party
# 'Votes', - This is RunnerUp Candidate Votes Count
header1 = ["PC No","PC Name","PC Type","Winning Candidate","Winning Color","Winning Party","Winning Votes",
          "RunnerUp Candidate","RunnerUp Color","RunnerUp Party","RunnerUp Votes","Margin"]
```

```python
# Reading the Data and putting in df after replacing headers(Task 1 - Replace Headers)
rows = data_table[1].findAll('tbody')[0].findAll('tr')[2:]
data_rows = [[td.get_text().strip() for td in tr.findAll('td')] for tr in rows]
df_web = pd.DataFrame(data_rows, columns=header1)
df_web.head(10)
```

```python
# Will Clear the Winning Color and RunnerUp Color Columns as I will not use them
# Task 2 - Format data into a more readable format
df_web = df_web.drop(["Winning Color","RunnerUp Color"], axis=1)
df_web.head(10)
```

```python
# Task 3 - Identify outliers and bad data
# For our analysis, PC Name, Winning Candidate is mandatory
# So checking if any of the values is null
print("Is Any Candidate Name Null - %r " % df_web['Winning Candidate'].isnull().values.any())
print("Is Any PC Name Null - %s " % df_web['PC Name'].isnull().values.any())

#Task 4 - Find duplicates
# To check for duplicates, PC Name Should be Unique (Generally it should be combination of PC Name and No)
print("Is Any PC Name duplicate ? - {}".format(any(df_web['PC Name'].duplicated())))
print("Is Any PC Name duplicate ? - {}".format(any(df_web['PC Name'].duplicated())))
# duplicate Rows with PC Name Only
duplicateRowsDF = df_web[df_web.duplicated(['PC Name'])]
display(duplicateRowsDF)
# Checking the Combination of PC Name and PC No
duplicateRowsDF = df_web[df_web.duplicated(['PC Name','PC No'])]
display(duplicateRowsDF)
# Since the Combination check returns no values we are good
```

```python
# Task 5 - Converting Votes Counts columns to int and also replace commnas from the string
import string

# first replacing commas
df_web['Winning Votes'] = df_web['Winning Votes'].str.translate(str.maketrans('','',string.punctuation))
df_web['RunnerUp Votes'] = df_web['RunnerUp Votes'].str.translate(str.maketrans('','',string.punctuation))
df_web['Margin'] = df_web['Margin'].str.translate(str.maketrans('','',string.punctuation))
df_web
```

```python
# Task 6 - Changing Votes Counts and Margin columns as integer types
df_web['Winning Votes'] = df_web['Winning Votes'].astype(int)
df_web['RunnerUp Votes'] = df_web['RunnerUp Votes'].astype(int)
df_web['Margin'] = df_web['Margin'].astype(int)
df_web
```

```python
# Task 7 - Checking if the Margin Calculated is same as present in data
# Setting the new column validity value as True if its correct
df_web["Validity"] = (df_web['Winning Votes'] - df_web['RunnerUp Votes'] == df_web['Margin'])
df_web
```

```python
# Task 8 - Checking if any data is invalid as per this new Validity column
print("Is Any Data Invalid ? - {}".format(any(df_web['Validity'] == False)))
```
